```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Local Minimum

Define the local min function and set up the grid.

```
def f(x, y):
  return x ** 2 + y ** 2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

Display the 2D contour map and label the min point.

```
plt.contourf(X, Y, Z, 20, cmap='coolwarm')
plt.plot(0, 0, 'go')
plt.colorbar()
```

Display the 3D contour map and label the min point.

```
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='coolwarm')
ax.scatter3D(0, 0, f(0,0), marker='o', c='g')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(60)
```

## Local Maximum

Define the local max function and update Z data.

```
def g(x, y):
  return -(x ** 2 + y ** 2)

Z = g(X, Y)
```

Display the 2D contour map and label the max point.

```
plt.contourf(X, Y, Z, 20, cmap='coolwarm')
plt.plot(0, 0, 'go')
plt.colorbar()
```

Display the 3D contour map and label the max point.

```
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='coolwarm')
ax.scatter3D(0, 0, f(0,0), marker='o', c='g')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(50)
```

## Saddle Point

Define saddle point function and update Z data.

```
def h(x, y):
  return x ** 2 - y ** 2

Z = h(X, Y)
```

Display the 2D contour map and label the saddle point.

```
plt.contourf(X, Y, Z, 20, cmap='coolwarm')
plt.plot(0, 0, 'go')
plt.colorbar()
```

Display the 3D contour map and label the saddle point.

```
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='coolwarm')
ax.scatter3D(0, 0, f(0,0), marker='o', c='g')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(40)
```
